```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df_long=pd.read_csv('datasets/rainfall.csv',encoding='utf-8')
```

```python
df_long.columns
```

```python
print(df_long.head())
```

```python
daily_rainfall = df_long['rainfall']
```

```python
plt.figure(figsize=(15, 6))
daily_rainfall.plot(title='Total Daily Rainfall Across All Locations')
plt.xlabel('Date')
plt.ylabel('Total Rainfall')
plt.grid(True)
plt.show()
```

```python
# You can also do this for monthly data to see seasonal patterns clearly
monthly_rainfall = df_long['rainfall'].resample('M').sum()
monthly_rainfall.plot(kind='bar', figsize=(15, 6), title='Total Monthly Rainfall')
plt.show()
```
