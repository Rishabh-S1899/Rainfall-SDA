#  Is the climate changing?

```python
import pandas as pd
import numpy as np
import pymannkendall as mk
import matplotlib.pyplot as plt
import seaborn as sns

# --- 2. SETUP ---
START_YEAR = 1992
NUM_YEARS = 33

# --- 2. Load the Data ---
print("Loading dataset...")
df = pd.read_csv('weather_coordinates.csv', index_col=0)

# Separate location info
location_data = df[['longitude', 'latitude']].copy()
rainfall_data = df.drop(columns=['longitude', 'latitude'])

# --- 3. Preprocess: Convert Hourly â†’ Yearly ---
print("Preprocessing data: Aggregating into yearly totals...")

# Clean column names: remove trailing decimals like ".33"
cleaned_cols = [col.split('.')[0] for col in rainfall_data.columns]

current_year = START_YEAR
parsed_dates = []
last_month_day = None

for col in cleaned_cols:
    # --- Handle 24:00 issue ---
    if "24:00" in col:
        # Split into date + dummy hour
        day_month = col.split()[0]   # e.g. "01-01"
        hour = col.split()[1]        # "24:00"
        
        # Convert to datetime (00:00 of same day)
        dt = pd.to_datetime(day_month + " 00:00", format="%d-%m %H:%M", errors="coerce")
        if dt is pd.NaT:
            raise ValueError(f"Could not parse column name: {col}")
        
        # Shift to next day
        dt = dt + pd.Timedelta(days=1)
    else:
        dt = pd.to_datetime(col, format="%d-%m %H:%M", errors="coerce")
        if dt is pd.NaT:
            raise ValueError(f"Could not parse column name: {col}")
    
    # --- Detect year rollover ---
    if last_month_day and dt.strftime('%m-%d %H:%M') < last_month_day:
        current_year += 1
    
    # Apply current year
    dt = dt.replace(year=current_year)
    parsed_dates.append(dt)
    
    last_month_day = dt.strftime('%m-%d %H:%M')

# Replace rainfall_data column headers with parsed datetimes
rainfall_data.columns = parsed_dates

# Aggregate by year
yearly_rainfall = rainfall_data.groupby(rainfall_data.columns.year, axis=1).sum()

print("Yearly rainfall prepared:")
print(yearly_rainfall.head())
# --- 4. Mann-Kendall Trend Analysis ---
print("\nPerforming Mann-Kendall Trend Analysis...")
results = []

for _, row in yearly_rainfall.iterrows():
    ts = row.values
    if np.var(ts) > 0:
        test_result = mk.original_test(ts, alpha=0.05)
        results.append([
            test_result.trend,
            test_result.p,
            test_result.s,
            test_result.slope
        ])
    else:
        results.append(['no trend', 1.0, 0.0, 0.0])

trend_results = pd.DataFrame(
    results,
    columns=['trend', 'p_value', 's_statistic', 'sen_slope'],
    index=location_data.index
)

# --- 5. Combine Results ---
final_df = location_data.join(trend_results)
final_df['significance'] = np.where(final_df['p_value'] < 0.05, 'Significant', 'Not Significant')

print("\nTrend analysis complete. Sample results:")
print(final_df.head())

# --- 6. Visualization ---
print("\nGenerating trend map...")

plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(12, 10))

sns.scatterplot(
    data=final_df,
    x='longitude',
    y='latitude',
    hue='trend',
    hue_order=['increasing', 'decreasing', 'no trend'],
    palette={'increasing': 'royalblue', 'decreasing': 'crimson', 'no trend': 'lightgray'},
    style='significance',
    markers={'Significant': 'X', 'Not Significant': 'o'},
    s=100,
    alpha=0.8,
    ax=ax
)

ax.set_title('Act 1: Spatiotemporal Trend of Rainfall', fontsize=18, weight='bold')
ax.set_xlabel('Longitude', fontsize=12)
ax.set_ylabel('Latitude', fontsize=12)
ax.legend(title='Trend (p < 0.05)', title_fontsize=13, fontsize=11, markerscale=1.5)
ax.set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.savefig("act1_trend_map.png", dpi=300)
plt.show()
```

# Clustering

```python
import pandas as pd
import numpy as np
import pymannkendall as mk
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# --- 1. SETUP ---

START_YEAR = 1992
N_CLUSTERS = 3

# --- 2. Load and Preprocess Data ---
print("Loading and preprocessing the dataset...")
df = pd.read_csv('weather_coordinates.csv', index_col=0)

location_data = df[['longitude', 'latitude']].copy()
rainfall_data = df.drop(columns=['longitude', 'latitude'])

cleaned_cols = [col.split('.')[0] for col in rainfall_data.columns]

current_year = START_YEAR
parsed_dates = []
last_month_day = None

for col in cleaned_cols:
    # --- Handle 24:00 issue ---
    if "24:00" in col:
        # Split into date + dummy hour
        day_month = col.split()[0]   # e.g. "01-01"
        hour = col.split()[1]        # "24:00"
        
        # Convert to datetime (00:00 of same day)
        dt = pd.to_datetime(day_month + " 00:00", format="%d-%m %H:%M", errors="coerce")
        if dt is pd.NaT:
            raise ValueError(f"Could not parse column name: {col}")
        
        # Shift to next day
        dt = dt + pd.Timedelta(days=1)
    else:
        dt = pd.to_datetime(col, format="%d-%m %H:%M", errors="coerce")
        if dt is pd.NaT:
            raise ValueError(f"Could not parse column name: {col}")
    
    # --- Detect year rollover ---
    if last_month_day and dt.strftime('%m-%d %H:%M') < last_month_day:
        current_year += 1
    
    # Apply current year
    dt = dt.replace(year=current_year)
    parsed_dates.append(dt)
    
    last_month_day = dt.strftime('%m-%d %H:%M')

# Replace rainfall_data column headers with parsed datetimes
rainfall_data.columns = parsed_dates

# Aggregate into yearly totals
yearly_rainfall = rainfall_data.groupby(by=rainfall_data.columns, axis=1).sum()

# --- 3. Feature Engineering ---
print("\nEngineering features for clustering...")

# Feature 1: Mean annual rainfall
mean_rainfall = yearly_rainfall.mean(axis=1)

# Feature 2: Variability
std_dev_rainfall = yearly_rainfall.std(axis=1)

# Feature 3: Trend (Senâ€™s slope)
slopes = []
for index, row in yearly_rainfall.iterrows():
    ts = row.values.astype(float)
    if np.var(ts) > 0:
        slope_res = mk.sens_slope(ts)
        slopes.append(slope_res.slope)
    else:
        slopes.append(0.0)

features_df = pd.DataFrame({
    'mean_rainfall': mean_rainfall,
    'std_dev_rainfall': std_dev_rainfall,
    'sen_slope': slopes
}, index=location_data.index)

# Standardize
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features_df)

# --- 4. Elbow Method ---
print("\nDetermining optimal number of clusters using the Elbow Method...")
inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_range)
plt.grid(True)
plt.savefig("act2_elbow_plot.png", dpi=300)
plt.show()

print(f"--> Based on the elbow plot, we are proceeding with k={N_CLUSTERS}.")

# --- 5. K-Means Clustering ---
print(f"\nPerforming K-Means clustering with k={N_CLUSTERS}...")
kmeans = KMeans(n_clusters=N_CLUSTERS, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(scaled_features)

final_df = location_data.copy()
final_df['cluster'] = cluster_labels
features_df['cluster'] = cluster_labels

# --- 6. Cluster Analysis ---
print("\nCluster summary (mean feature values):")
cluster_summary = features_df.groupby('cluster').mean()
print(cluster_summary)

print("\nGenerating cluster map...")
plt.figure(figsize=(12, 10))
sns.scatterplot(
    data=final_df,
    x='longitude',
    y='latitude',
    hue='cluster',
    palette='viridis',
    s=100,
    alpha=0.8,
    legend='full'
)
plt.title('Act 2: The Climate Zones of Northern India', fontsize=18, weight='bold')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Cluster', title_fontsize=13, fontsize=11)
plt.gca().set_aspect('equal', adjustable='box')
plt.tight_layout()
plt.savefig("act2_cluster_map.png", dpi=300)
plt.show()
```
